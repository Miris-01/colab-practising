```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#DATOS DE LOS FAMILIARES

data = {
    "Familiar": [
        "Mamá",
        "Papá",
        "Hermana",
        "Yo",
        "Abuela",
        "Abuelo",
        "Tío 1",
        "Tía 1",
        "Primo 1",
        "Primo 2",
    ],
    "Edad": [63, 67, 34, 32, 84, 86, 65, 63, 35, 8],
    "Género": [
        "Femenino",
        "Masculino",
        "Femenino",
        "Femenino",
        "Femenino",
        "Masculino",
        "Masculino",
        "Femenino",
        "Masculino",
        "Femenino",
    ],

    "Horas_Movil": [14, 12, 28, 25, 4, 3, 18, 20, 15, 5],
}
```

```python
df = pd.DataFrame(data) #creamos el dataframe con los datos que hemos declarado previamente
```

```python
#unos pocos de prints para ver conceptos estadísticos:

print(f"Media de edad: {df['Edad'].mean():.2f} años")

print(f"Media de horas en el móvil {df['Horas_Movil'].mean():.2f} horas")

age_sorted = sorted(df['Edad'])
print(f"Mediana Q2: {df['Edad'].median():.1f} años") #mediana: valor central de un conjunto de datos

print(f"Moda: {df['Edad'].mode()[0]:.2f} años")#el número de veces que se repite un valor
print(df['Edad'].value_counts())

#percentiles y cuartiles: medidas de posición estadística que dividen un conjunto de datos ordenados en partes porcentualmente iguales.

#print(f"Percentil 25: {np.percentile(df['Edad'], 25)}")#el 25% esta por debajo de este resultado Q1
#print(f"Percentil 50: {np.percentile(df['Edad'], 50)}")#el 50% esta por debajo de este resultado Equivale al Q2 y a la mediana
#print(f"Percentil 75: {np.percentile(df['Edad'], 75)}")#el 75% esta por debajo de este resultado Q3
#print(f"Percentil 90: {np.percentile(df['Edad'], 90)}")#el 90% esta por debajo de este resultado

#MEDIDAS DE DISPERSION
#RANGO
rango = max(df['Horas_Movil']) - min(df['Horas_Movil'])
print(f"Rango de horas con el movil {rango}")

#RANGO INTERCUARTILICO: representa el 50% central de los datos
rango_in = np.percentile(df['Edad'], 75) - np.percentile(df['Edad'], 25) #Q3 - Q1
#print(f'Rango intercuartil: {rango_in}')
#detectar outliners
lim_inf = df['Edad'].quantile(0.25)
#print('Limite inferior',lim_inf)
lim_sup = df['Edad'].quantile(0.75)
#print('Limite superior',lim_sup)
#VARIANZA: mide la dispersion promedio al cuadrado respecto a la media
varianza = np.var(df["Horas_Movil"], ddof=1)
print(f'varianza: {varianza:.2f}')

```

**EMPEZAMOS CON LOS GRAFICOS**

```python
#Cargamos configuracion visual inicial

sns.set_theme(
    style = 'darkgrid',
    palette = 'muted',
    context = 'talk',
    rc={
        'figure.figsize': (8,5),
        'axes.titlesize': 16,
        'axes.labelsize': 12
    },
)
```

```python
sns.histplot(data=df, x='Edad',bins=6, kde=True)
plt.title('Distribucion de edades en mi familia')
plt.show()
#No hay nadie en los 50!!!!
```

```python
#SCATTERPLOT
sns.scatterplot(x='Edad', y='Horas_Movil', hue='Género', data=df, s=150)
plt.title('Relacion: edad VS uso del movil')
plt.legend(bbox_to_anchor=(1.05, 1), loc= 'upper left', title='Género')
plt.show()
```

```python
#BOXPLOT (muy importante a nivel empresarial)
sns.boxplot(data=df, x='Género', y='Horas_Movil')
plt.title('Comparativa del Uso del Móvil por Género')
plt.show()
```
